```python
#Removing Outliers
Q1 = df_balancedBank.quantile(0.25)
Q3 = df_balancedBank.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df_balancedBank = df_balancedBank[~((df_balancedBank < (Q1-1.5 * IQR)) |(df_balancedBank > (Q3 + 1.5 * IQR))).any(axis=1)]
df_balancedBank.shape
```
